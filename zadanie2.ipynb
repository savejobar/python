```python
a0 = 1
d = 5
q = 7
n = 5
for i in range (1,n+1):  #Arifmetic
    Sn_ar = (2*a0+d*(i-1))*i*0.5
    print("n =",i,"\tSum = ",Sn_ar)
print()
for i in range (1,n+1):  #Geom
    Sn_geom = a0*(q**(i)-1)/(q-1)
    print(f"n = {i}\tSum = {Sn_geom}")

```
