```python
import requests
from http import HTTPStatus         
from bs4 import BeautifulSoup

def get_page(url: str):
    response = requests.get(url)
    if response.status_code != HTTPStatus.OK:
        raise Exception(f'Invalid http status. got {response.status_code}')
    soup = BeautifulSoup(response.text,'html.parser')
    
    return soup.text
```

```python
page_url = 'https://service-bts.ru/'

with open ('file_text.txt', 'w', encoding = 'utf-8') as f:
    f.write(get_page(page_url))
```

```python
import re

def seek_of_names(text_of_file: str):
    if not isinstance(text_of_file, str):
        raise ValueError('file must be a string')
    text_of_file = re.sub(r"<.*?>", " ", text_of_file)
    text_of_file = re.sub(r"\s+", " ", text_of_file)
    words = re.findall(r'\b[A-ZА-ЯЁ][a-zа-яёA-ZА-ЯЁ]+\b', text_of_file)
    result = "; ".join(words)
    return result
    
```

```python
with open ('file_text.txt', encoding = 'utf-8') as f:
    text_of_file = f.read()
print(seek_of_names(text_of_file))

with open ('file_test_of_names.txt', 'w', encoding = 'utf-8') as f:
    f.write(seek_of_names(text_of_file))
```

```python
def ununique_names(filename: str):
    if not isinstance(filename, str):
        raise ValueError('file must be a string')
    with open (filename, encoding = 'utf-8') as f:
        text_of_file = f.read()
    text_of_file = re.sub(r"<.*?>", " ", text_of_file)
    text_of_file = re.sub(r"\s+", " ", text_of_file)
    words = re.findall(r'\b[A-ZА-ЯЁ][a-zа-яёA-ZА-ЯЁ]*\b', text_of_file)
    aux = {}
    for el in words:
        if el in aux:
            aux[el] += 1
        else:
            aux[el] = 1

    return aux
```

```python
ununique_names('file_test_of_names.txt')
```

```python
def count_names(dictionary: dict):
    dictionary = sorted(dictionary.items(), reverse = True, key = lambda item: item[1])
    for key, value in dictionary:
        print(f'{key} = {value}')
        break
```

```python
count_names(ununique_names('file_test_of_names.txt'))
```
